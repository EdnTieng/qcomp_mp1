Import the modules

```python
import numpy as np
from scipy.optimize import minimize
```

Define values for theta, phi, and lambda

```python
input = [0, 0, 0] #theta, phi, and lambda respectively
```

Define the vectors

```python
# To be used inside the loss function
def U(theta, phi, lambdaaa):
    return np.array([
        np.cos(theta / 2), # U[0][0]
        -np.exp(1j * lambdaaa) * np.sin(theta / 2), # U[0][1]
        np.exp(1j * phi) * np.sin(theta / 2), # U[1][0]
        np.exp(1j * (phi + lambdaaa)) * np.cos(theta / 2) # U[1][1]
    ])

# H vector
H = np.array([1, 1, 1, -1]) / np.sqrt(2)
```

Define the loss function

```python
# Loss Function (Squared Euclidean Distance)
def loss(values):
    theta, phi, lambdaaa = values
    return np.sum(np.abs(U(theta, phi, lambdaaa) - H)**2)
```

Call the minimize function

```python
# To minimize the loss function
result = minimize(loss, input, method='BFGS')

# Get output values from minimized loss
output = result.x

# Print the results
print(f"Input theta: {input[0]}")
print(f"Input phi: {input[1]}")
print(f"Input lambda: {input[2]}")
print("\n")
print(f"Output theta: {output[0]}")
print(f"Output phi: {output[1]}")
print(f"Output lambda: {output[2]}")
```
