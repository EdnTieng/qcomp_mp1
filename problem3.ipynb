```python
import numpy as np
from scipy.optimize import minimize

# Rotation around Z axis
def Rz(theta):
    theta = float(theta)
    return np.array([
        [np.cos(theta), -np.sin(theta), 0],
        [np.sin(theta),  np.cos(theta), 0],
        [0,              0,             1]
    ])

# Target vector Z
Z = np.array([0, 0, 1])

# Loss function
def loss(theta):
    rotated = Rz(theta) @ Z
    return np.linalg.norm(rotated - Z)**2

# optimization
result = minimize(loss, x0=[1.0])

print("Optimal theta:", result.x[0])
print("Loss:", result.fun)
```
